```python
import spacy
nlp = spacy.load("en_core_web_sm")
with open("data/wiki_us.txt", "r") as f:
    text = f.read()
```

```python
print(text)
```

```python
doc = nlp(text)

for token in doc[:10]:
    print(token)
```

```python

sentences = list(doc.sents)
print(sentences[0])

token2 = sentences[0][2]
print("FOR ", token2)
print(token2.text)
print(token2.left_edge)
print(token2.right_edge)
print(token2.ent_type_)
print(token2.ent_iob_)
print(token2.lemma_)
print(token2.pos_)
print(token2.dep_)
print(token2.lang_)

token3 = sentences[0][12]
print("FOR ", token3)
print(token3.lemma_)
print(token3.morph)
```

```python
print("================================================")

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
print("================================================")
# text = "Mike enjoys playing football."
text = "Create a variable called A and save 12 in it."
print(text)
doc2 = nlp(text)

for token in doc2:
    print(token.text, token.pos_, token.dep_, token.tag_)
    if (token.text == "it"):
        print(token.text, token.ent_kb_id_)
```
